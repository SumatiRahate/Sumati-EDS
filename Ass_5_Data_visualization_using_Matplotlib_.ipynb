```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv('/content/Book1.csv')



# Assuming you have a dataset stored in a pandas DataFrame
df = pd.DataFrame(dataset)

# Convert 'Company' column into a list
company_list = df['Companies'].tolist()

# Convert 'Product' column into a list
product_list = df['Product'].tolist()

#Convert 'Supplier' column into a list
Supplier_list = df['Supplier'].tolist()

#Convert 'Consumer' column into a list
consumer_list = df['Consumer'].tolist()

#Convert 'Gender' column into a list
gender_list = df['Gender'].tolist()

#Convert 'Residence' column into a list
residence_list = df['Residance'].tolist()

# Print the converted lists
print(company_list)
print(product_list)
print(Supplier_list)
print(consumer_list)
print(gender_list)
print(residence_list)





# Graph 1: Scatter plot of residence vs. supplier
residence_supplier = list(zip(residence_list, Supplier_list))

plt.figure(figsize=(8, 5))
for residence in set(residence_list):
    x = [pair[1] for pair in residence_supplier if pair[0] == residence]
    y = [pair[0] for pair in residence_supplier if pair[0] == residence]
    plt.scatter(x, y, label=residence)

plt.xlabel('Supplier')
plt.ylabel('Residence')
plt.title('Residence vs. Supplier')
plt.legend()
plt.show()





# Graph 2 : Count the occurrences of each product
product_counts = {}
for product in product_list:
    product_counts[product] = product_counts.get(product, 0) + 1

# Graph 3 : Bar graph of product counts
plt.bar(product_counts.keys(), product_counts.values())
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Product Counts')
plt.show()

# Graph 4 : Pie chart of gender distribution
gender_counts = {}
for gender in gender_list:
    gender_counts[gender] = gender_counts.get(gender, 0) + 1

plt.pie(gender_counts.values(), labels=gender_counts.keys(), autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()


# Graph 5 :Bar graph of product count per company
plt.figure(figsize=(8, 6))
plt.bar(company_list, product_list)
plt.xlabel("Companies")
plt.ylabel("Product")
plt.title("Product Count per Company")
plt.xticks(rotation=45)
plt.show()
```
